```python
# Dependencies
import csv
import json
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
```

```python
data_file = "UR_KAN_pubs.csv"
```

```python
data_file_pd = pd.read_csv(data_file)
data_file_pd["Publication Year"]= data_file_pd["Publication Year"].fillna(0.0).astype(int)
data_file_pd.head()
```

```python
UR_pubs = data_file_pd[["Authors", "Journal Title", "Publication Year"]]
UR_pubs.head()
```

```python
journal_grouping = data_file_pd.groupby(["Journal Title"]).count()[["Title"]]
journal_grouping
```

```python
journal_sort = journal_grouping.sort_values("Title", ascending=False)
journal_sort.head()
```

```python
yearly_publications = data_file_pd.groupby(["Publication Year"]).count()[["Title"]]
yearly_publications
```

```python
yearly_publications_per_journal = data_file_pd.groupby(["Publication Year","Journal Title"]).count()[["Title"]]
yearly_publications_per_journal
```
